## 3. Bicycle Counts

```python
bike_count = pd.read_csv('bike_traffic_data.csv')
bike_count['Date'] = pd.to_datetime(bike_count['Date'])
bike_count.head(10)
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=bike_count, x='Date', y='Fremont Bridge West Sidewalk')
plt.show()
```
